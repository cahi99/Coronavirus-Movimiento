```
'''Este índice de contagio nos ayudará a medir el incremento porcentual en base
al número de contagiados.
En conjunto con el entrenamiento con los datos de movimiento
se logrará calcular qué tipo de actividad incrementa el riesgo a qué porcentaje.
===
El resultado final se consigue de multiplicar el número para la cantidad de contagiados actuales.
(Este último no es resultado de la máquina de entrenamiento, sinó de un cálculo matemático simple)'''
temp=[]
indice_contagio=[]
#para los positivos
for i in range(len(Datos['Positivos'].values)):
  map_object = map(int, Datos['Positivos'].values[i])#prs
  l1 = list(map_object)#prs Lista de contagios totales al día (Base)
  if (i+1)!=len(Datos['Positivos'].values):#validacioni para comparar listas
    map_object2 = map(int, lista_positivos[i])#prs
    l2 = list(map_object2)#prs Lista de contagios nuevos adquiridos al día (Dividendo)
    temp=[x2 / x1 if (x1!=0)>0 else x2/1 for (x1, x2) in zip(l1, l2)]
    indice_contagio.append(temp)
len(indice_contagio)
```

```
def indice_movimiento(*args):
  temp=[]
  indice_movimiento=[]
  #para los positivos
  for i in range(len(np.asarray(args[0].values))):
    if (i+1)!=len(np.asarray(args[0].values)):
      l1=args[0].values[i]
      l2=args[0].values[i+1]
      if l1 < 0:
        l1=l1*-1
      if l2 < 0:
        l2=l2*-1
      temp=l2 / l1 if l1!=0 else l2/1 
      indice_movimiento.append(temp)
  return indice_movimiento
```

```
ListaPaises={}

def InsertarLista(Pais, dir):
  #Transformación del formato de las fechas.
  XD=pd.read_csv(dir, sep=",", header=0)
  import datetime
  for i in range(len(XD['ds'])):
    j=datetime.datetime.strptime(XD['ds'][i], '%Y-%m-%d').strftime('%m/%d/%y')
    XD['ds'][i]=j
  XDD=XD.groupby('ds')
  fechas=[fecha for fecha in XDD['ds'].first()]
  #Fechas para realizar el promedio...
  #Promedio...
  lista_prom=[]
  from functools import reduce
  for k in fechas: #Por cada día...
    lista_temporal1=[]
    lista_temporal2=[]
    for i in XD.values: #Por cada fila...
      if i[0]==k:
        lista_temporal1.append(abs(i[5]))
        lista_temporal2.append(abs(i[6]))
    promediado1=reduce(lambda x,y: x+y,lista_temporal1)/len(lista_temporal1)
    promediado2=reduce(lambda x,y: x+y,lista_temporal2)/len(lista_temporal2)
    k=datetime.datetime.strptime(k, '%m/%d/%y').strftime('%d/%m/%y')
    lista_prom.append([k,promediado1,promediado2])
  global ListaPaises 
  ListaPaises[Pais]=pd.DataFrame(lista_prom, columns=["Fecha","Mov1","Mov2"])
  print('País ingresado.')

```
